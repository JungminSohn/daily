# sampling

```python
import pickle
import numpy as np
import pandas as pd

with open('model3_input.pickle','rb') as f:
    data=pickle.load(f)
```

```python
data
```

```python
notdrop = data[data.철회자==0]
drop = data[data.철회자==1]    
```

```python
notdrop
```

```python
drop
```

drop(6599개)와 똑같은 수의 notdrop 샘플을 10번 뽑는다. (1:1 데이터셋 10개)

```python
sample_set=[]
for t in range(10):
    sample_set.append(notdrop.sample(len(drop)).append(drop))
```
